```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

# 5K GEN 10 POP AND BEST MATES WITH EVERYONE / NO PREDATION

```python
#Loading data...
log1 = pd.read_csv('log15k10B.csv',encoding='utf-8')
log2 = pd.read_csv('log25k10B.csv',encoding='utf-8')
```

```python
x1 = log1['Generation']
y1 = log1['Fitness']
x2 = log2['Generation']
y2 = log2['Fitness']

plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.plot(x1,y1)
plt.plot(x2,y2)
```
