```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

# 5K GEN 10 POP AND BEST MATES WITH EVERYONE / NO PREDATION

```python
#Para 0.0001 segundos
log1 = pd.read_csv('5kB11.csv',encoding='utf-8')
log2 = pd.read_csv('5kB21.csv',encoding='utf-8')
x1 = log1['Generation']
y1 = log1['Fitness']
x2 = log2['Generation']
y2 = log2['Fitness']

plt.suptitle('Best/No Predation', fontsize=14, fontweight='bold')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.plot(x1,y1,'b', label = 'Population 1',linewidth=2.0)
plt.plot(x2,y2,'m', label = 'Population 2',linewidth=2.0)
plt.legend()
```
